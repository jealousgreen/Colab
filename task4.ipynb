```python
import numpy as np
```

```python
temp = np.array([[1, 2, 1], [-2, 1, 1]])
A = temp.transpose()
print ("Матрица A:\n", A)
B = np.array([[3, 1], [-1, 0], [2, 4]])
print ("Матрица B:\n",B)
C = np.array([[-2, 2], [1, -1], [1, 2]])
print ("Матрица C:\n",C)
```

```python
r1=A+(2*B)
print ("A+2B:\n", r1)

r2=B+(2*C)
print ("B+2C:\n", r2)
```
